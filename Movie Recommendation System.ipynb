```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv("movies.csv")
```

```python
movies_df.head()
```

```python
# explain the below libs
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# selecting the relevant features for recommendation
selected_features = ["genres","keywords","tagline","director","cast"]
```

```python
# replacing the null values with null string

for feature in selected_features:
    movies_df[feature] = movies_df[feature].fillna("")
```

```python
# combining all the five selected features

combined_features = movies_df["genres"]+ movies_df["keywords"]+ movies_df["tagline"]+ movies_df["cast"]+ movies_df["director"]
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# Cosine Similarity
```

```python
# getting the similarity score using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
# getting the movie name from the user
movie_name = input(" Enter your favourite movie name : ")
```

```python
# creating a list with all the movie names given in the movie dataset
```

```python
list_of_all_titles = movies_df["title"].tolist()
```

```python
# finding the close match movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
```

```python
find_close_match
```

```python
close_match = find_close_match[0]
```

```python
# finding the index of the movie with title
index_of_the_movie = movies_df[movies_df.title == close_match]["index"].values[0]
```

```python
index_of_the_movie
```

```python
# getting the list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_score = sorted(similarity_score, key= lambda x:x[1], reverse = True)
```

```python
sorted_similar_score
```

```python
# print the name of similar movies based on the index
print("Movies suggested for you: \n")
i = 1
for movie in sorted_similar_score:
    index = movie[0]
    title_from_index = movies_df[movies_df.index==index]["title"].values[0]
    if (i<30):
        print(i, '.',title_from_index)
        i+=1
```
